```python
# Create text file with list of blog post sites
import os

def write_to_txt_file(list_of_texts):
    isSuccess = True
    try:
        if os.path.exists('./' + list_of_blogs_urls_for_scrapping):
            f = open('./' + list_of_blogs_urls_for_scrapping, "a+")
        else:
            f = open('./' + list_of_blogs_urls_for_scrapping,"w+")
        for text in list_of_texts:
             f.write(text + '\n')
        f.close()
    except Exception as e:
        print(e)
        isSuccess = False
    finally: 
        return isSuccess 
```

```python
import time, random

def get_google_search_results(query, start_id = 0, page_contain_link = True):
    links = []
    try:
        while page_contain_link:
            page_contain_link = False

            # Google search
            url = 'http://www.google.com/search?q={query}&oq={query}&start={start_id}'.format(query = query, start_id = start_id)
            temp_links = []
            
            # Extracting result urls from google searches
            html_page = requests.get(url).text
            soup = BeautifulSoup(html_page, "html.parser")
            
            for link in soup.findAll('a', attrs={'href': re.compile(r"http")}):
                exp_0 = 'url?q='
                exp_1 = '&amp;'
                link = str(link)
                if exp_0 in link:
                    link = link.split(exp_0)[1].split(exp_1)[0]
                    if 'google' not in link: # Include links that do not contain 'google'
                        temp_links.append(link)
#                         print(link)
            # Continue search for subsequent result pages
#             if temp_links != []:
#                 page_contain_link = True
#                 start_id = start_id + 10
            links.extend(temp_links)
            write_to_txt_file(temp_links)
            print(links)
            # Time Delay
            time.sleep(random.randrange(5,10))
            return links
    except Exception as e:
        print("Error")
        print(e)
        return e
```

```python
# Replace text based on key, value in dictionary
# Inputs:
#    s              : String
#    text_to_replace: Dict   
# Outputs:
#    s              : String
def clean_text(s, text_to_replace):
    for k,v in text_to_replace.items():
        s = s.replace(k, v)
    return s

# Remove tag elements based on the specific start and end tags
# Inputs:
#    s        : String
#    start_tag: String (E.g. '<a')
#    end_tag  : String (E.g. '>')
# Outputs:
#    s        : String
def remove_html_element(s, start_tag, end_tag):
    full_para = ''
    if '<' in s:
#         print('Start Tag Found')
        split_first_part, split_second_part = s.split(start_tag, 1)
        split_second_part = split_second_part.split(end_tag, 1)[1]
        s = split_first_part + split_second_part
        s = s.replace('  ', ' ')
    return s
```

```python
def extract_date(url):
    try:
        date = re.findall(r'/(\d{4})/(\d{1,2})/(\d{1,2})/', url)
        if not date:
            date = re.findall(r'/(\d{4})/(\d{1,2})/', url)[0]
            date = tuple([date[0], date[1], '01'])
        else:
            date = date[0]
        date = "-".join(date)
#         date = list(date)
    except Exception as e:
#         print(e)
        date = ''
    finally:
        return date
        
# url1= "https://www.washingtonpost.com/news/football-insider/wp/2016/09/02/odell-beckhams-fame-rests-on-one-stupid-little-ball-josh-norman-tells-author/"
# print(extract_date(url1))
```

```python
import time

text_to_replace = {
    '<b>': '', '</b>': '', '<p>': '', '</p>': '', '</a>': '', '<li>': '', '</li>': '', '<strong>': '', '</strong>': '',
    '<noscript>': '', '</noscript>': '', '<em>': '', '</em>': '', '</div>': '', '</span>': '', '</td>': '', '<br/>': '',
    '</img>': '', '&amp;': '&', '\n': '', '\r': ''
}
list_texts_para_should_not_contain = ['class=', 'Copyright']

domain_names_from_diff_sources = []
dates_from_diff_sources = []
urls_from_diff_sources = []
blog_names_from_diff_sources = []
texts_without_dates = []

start_time = time.time()

# Extract results from all search results
for link in all_urls:
    print(link)
    blog_date, blog_name, domain_name = link_text_preprocessing(link, text_to_replace, list_texts_para_should_not_contain)
    domain_names_from_diff_sources.append(domain_name)
    dates_from_diff_sources.append(blog_date)
    urls_from_diff_sources.append(link)
    blog_names_from_diff_sources.append(blog_name)
    
    if not blog_date: # Add texts without date to list
        texts_without_dates.extend(domain_name)
    print('\n')
    
all_paras_df = pd.DataFrame(zip(dates_from_diff_sources, blog_names_from_diff_sources, domain_names_from_diff_sources, urls_from_diff_sources), columns = ['date', 'blog_name', 'domain_name', 'url'])
print('Dataframe Conversion Elapsed Time:', time.time() - start_time, 's')
print('Total number of texts:', len(domain_names_from_diff_sources))
print('Number of texts without date:', len(texts_without_dates))
```
