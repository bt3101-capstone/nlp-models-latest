```python
# Create text file with list of blog post sites
import os

def write_to_txt_file(list_of_texts, filename):
    isSuccess = True
    try:
        if os.path.exists('./' + filename):
            f = open('./' + filename, "a+")
        else:
            f = open('./' + filename,"w+")
        for text in list_of_texts:
             f.write(text + '\n')
        f.close()
    except Exception as e:
        print(e)
        isSuccess = False
    finally: 
        return isSuccess 
```

```python
import time, random

def get_google_search_results(query, start_id = 0, page_contain_link = True):
    links = []
    try:
        while page_contain_link:
            page_contain_link = False

            # Google search
            url = 'http://www.google.com/search?q={query}&oq={query}&start={start_id}'.format(query = query, start_id = start_id)
            temp_links = []
            
            # Extracting result urls from google searches
            html_page = requests.get(url).text
            soup = BeautifulSoup(html_page, "html.parser")
            
            for link in soup.findAll('a', attrs={'href': re.compile(r"http")}):
                exp_0 = 'url?q='
                exp_1 = '&amp;'
                link = str(link)
                if exp_0 in link:
                    link = link.split(exp_0)[1].split(exp_1)[0]
                    if 'google' not in link: # Include links that do not contain 'google'
                        temp_links.append(link)
#                         print(link)
            # Continue search for subsequent result pages
#             if temp_links != []:
#                 page_contain_link = True
#                 start_id = start_id + 10
            links.extend(temp_links)
    
            # Adds list of blog urls to txt file
            write_to_txt_file(temp_links, list_of_blogs_urls_for_scrapping)
            print(links)
    
            # Time Delay
            time.sleep(random.randrange(5,10))
            return links
    except Exception as e:
        print("Error")
        print(e)
        return e
```

```python
def extract_date(url):
    try:
        date = re.findall(r'/(\d{4})/(\d{1,2})/(\d{1,2})/', url)
        if not date:
            date = re.findall(r'/(\d{4})/(\d{1,2})/', url)[0]
            date = tuple([date[0], date[1], '01'])
        else:
            date = date[0]
        date = "-".join(date)
#         date = list(date)
    except Exception as e:
#         print(e)
        date = ''
    finally:
        return date
        
# url1= "https://www.washingtonpost.com/news/football-insider/wp/2016/09/02/odell-beckhams-fame-rests-on-one-stupid-little-ball-josh-norman-tells-author/"
# print(extract_date(url1))
```

## Scrap from all Google searched URLs

```python
from datetime import date

# Extract texts from specified link and cleans results
# Inputs:
#    link                              : String
# Outputs:
#    all_paras                         : list
def link_text_preprocessing(link):
    all_paras = []
    blog_name, date_str = '', ''
    min_length_text = 20
    try:
        html_link = requests.get(link).text
        soup = BeautifulSoup(html_link, "html.parser")
        
        ### Domain Name Extraction
        domain_name = link.split('://')[1].split('/', 1)[0]
        
        ### Blog Name Extraction
        blog_name = soup.title.string
        blog_name = re.sub(r'[^\x00-\x7F]+',' ', blog_name)
        print(blog_name)
        
        ### Date Extraction
        date_str = extract_date(link)
        if date_str:
            print('First Method')
        else:
            contains_date_first_type = soup.find(lambda tag:tag.name=="script" and "datePublished" in tag.text)
            if contains_date_first_type:
                print('Second Method')
                element_txt = contains_date_first_type.text
                start_idx = element_txt.find('datePublished')
                date_str = element_txt[start_idx:].split('T', 1)[0].split('\"')[-1]
            else:
                contains_date_second_type = soup.find('time', attrs={'class': re.compile(r"entry-date")})
                if contains_date_second_type:
                    print('Third Method')
                    element_txt = str(contains_date_second_type)
                    start_idx = element_txt.find('datetime')
                    date_str = element_txt[start_idx:].split('T', 1)[0].split('\"')[-1]
                else:
                    print('Cannot find blog date')
            
        if date_str:
            year, month, day = date_str.split('-')
            
    except Exception as e:
        print(e)
        
    finally:
        print(date_str)
        return date_str, blog_name, domain_name
```

## Clean scraped data

```python
# Read in scraped data
scraped_data = pd.read_csv('.\scraped_data_full.csv', dtype = {'date': str, 'blog_name': str, 'domain_name': str, 'url': str})
scraped_data.dtypes

# Remove non-date
date_length = len('DD/MM/YYYY')
print(len(scraped_data))

# Text to replace
text_to_replace = {
    '<b>': '', '</b>': '', '<p>': '', '</p>': '', '</a>': '', '<li>': '', '</li>': '', '<strong>': '', '</strong>': '',
    '<noscript>': '', '</noscript>': '', '<em>': '', '</em>': '', '</div>': '', '</span>': '', '</td>': '', '<br/>': '',
    '</img>': '', '&amp;': '&', '\n': '', '\r': '', '\t': ''
}

# Function to replace text
def replace_text(text_to_replace, s):
    for k,v in text_to_replace.items():
        s = s.replace(k,v)
    return s

# Cleaning dataframe
cleaned_scraped_data = scraped_data.copy()
cleaned_scraped_data['date'] = cleaned_scraped_data['date'].apply(lambda x: str(x) if len(str(x)) <= date_length else '')
cleaned_scraped_data['date'] = cleaned_scraped_data['date'].apply(lambda x: x if x != 'nan' else '')
cleaned_scraped_data['blog_name'] = cleaned_scraped_data['blog_name'].apply(lambda x: str(x) if x != 'Not Acceptable!' else '')
cleaned_scraped_data['blog_name'] = cleaned_scraped_data['blog_name'].apply(lambda x: replace_text(text_to_replace, x))

cleaned_scraped_data
```

```python
print('Total Number of blog posts:', len(cleaned_scraped_data))
print('Total Number of blog posts without date:', len(cleaned_scraped_data[cleaned_scraped_data['date'] == '']))
print('Total Number of blog posts without blog name:', len(cleaned_scraped_data[cleaned_scraped_data['blog_name'] == '']))
print('Total Number of blog posts with missing field:', len(cleaned_scraped_data[(cleaned_scraped_data['date'] == '') & (cleaned_scraped_data['blog_name'] == '')]))
print('Total Number of blog posts with all fields:', len(cleaned_scraped_data) - len(cleaned_scraped_data[(cleaned_scraped_data['date'] == '') & (cleaned_scraped_data['blog_name'] == '')]))
```

```python
filepath = './' + 'cleaned_scraped_data_full'

# Writes to output csv 
cleaned_scraped_data.to_csv(filepath + '.csv', index=False)
```

```python
scraped_data_db_insertion = []
uniqueBlogUrlList = cleaned_scraped_data['domain_name'].unique()
for blogUrl in uniqueBlogUrlList:
    temp_df = cleaned_scraped_data[cleaned_scraped_data['domain_name'] == blogUrl]
    blogs = []
#     print(temp_df)
    for index, row in temp_df.iterrows():
#         print(row)
        blogs.append({'blogPost': row['url'], 'locations': [], 'spacyNer': [], 'awsNer':[], 'writtenAt': row['date']})
    scraped_data_db_insertion.append({'blogUrl': blogUrl, 'blogs' : blogs})
scraped_data_db_insertion
```
